```python
'''
MAP TRANSACTION DATA
This code processes transaction data from JSON files organized by state, year, and quarter.
'''

import pandas as pd
import os
import json

map_transState_path = r"D:\guvi project\PhonePay\pulse\data\map\transaction\hover\country\india\state"

map_transState_list = os.listdir(map_transState_path)
map_transState_list

map_trans_data = {'State':[], 'Year':[], 'Quarter':[], 'District':[], 'Transaction_count':[], 'Transaction_amount':[]}

for state in map_transState_list:
    years = os.listdir(f"{map_transState_path}/{state}")
    for year in years:
        quarter = os.listdir(f"{map_transState_path}/{state}/{year}")
        for qtr in quarter:
            path = f"{map_transState_path}/{state}/{year}/{qtr}"
            with open(path, 'r') as f:
                data = json.load(f)
            for i in data['data']['hoverDataList']:
                district = i['name']
                try:
                    count = i['metric'][0]['count']
                    amount = i['metric'][0]['amount']
                except (KeyError, IndexError, TypeError) as e:
                    # If any error, assign None
                    count = None
                    amount = None
                map_trans_data['State'].append(state)
                map_trans_data['Year'].append(year)
                map_trans_data['Quarter'].append(int(qtr.strip('.json')))
                map_trans_data['District'].append(district)
                map_trans_data['Transaction_count'].append(count)
                map_trans_data['Transaction_amount'].append(amount)
        
map_trans_df = pd.DataFrame(map_trans_data)
map_trans_df
```

```python
'''
FOR DATABASE INSERTION OF MAP TRANSACTION DATA
'''
map_trans_df.to_sql('map_transaction', engine, if_exists='replace', index=False)
print("✅ Data inserted into PostgreSQL successfully.")
```

```python
'''
MAP USER DATA
This code processes user data from JSON files organized by state, year, and quarter.
'''
import pandas as pd
import os
import json

map_userstate_path = r'D:\guvi project\PhonePay\pulse\data\map\user\hover\country\india\state'
map_userstate_list = os.listdir(map_userstate_path)
map_user_data = {
    'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Registered_Users': [], 'App_Opens': []
}

for state in map_userstate_list:
    years = os.listdir(f"{map_userstate_path}/{state}")
    for year in years:
        # print(year)
        quarter = os.listdir(f"{map_userstate_path}/{state}/{year}")
        for qtr in quarter:
            # print(f"{map_userstate_path}/{state}/{year}/{qtr}")
            path = f"{map_userstate_path}/{state}/{year}/{qtr}"
            with open(path, 'r') as f:
                data = json.load(f)

            hover_data = data['data']['hoverData']
            for district, metrics in hover_data.items():
                try:
                    registered = metrics.get('registeredUsers', None)
                    opens = metrics.get('appOpens', None)
                except (KeyError, IndexError, TypeError) as e:
                    registered = None
                    opens = None
                
                map_user_data['State'].append(state)
                map_user_data['Year'].append(year)
                map_user_data['Quarter'].append(int(qtr.strip('.json')))
                map_user_data['District'].append(district)
                map_user_data['Registered_Users'].append(registered)
                map_user_data['App_Opens'].append(opens)
            
# Convert to DataFrame
map_user_df = pd.DataFrame(map_user_data)
map_user_df
```

```python
'''
FOR DATABASE INSERTION OF MAP USER DATA
'''
map_user_df.to_sql('map_user', engine, if_exists='replace', index=False)
print("✅ User data inserted into PostgreSQL successfully.")
```

```python
'''
MAP INSURANCE DATA
This code processes insurance data from JSON files organized by state, year, and quarter.
'''

import pandas as pd
import os
import json

map_insstate_path = r'D:\guvi project\PhonePay\pulse\data\map\insurance\country\india\state'
map_insstate_list = os.listdir(map_insstate_path)

map_ins_data = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'District': [],
    'Latitude': [],
    'Longitude': [],
    'Metric': []
}
for state in map_insstate_list:
    years = os.listdir(f"{map_insstate_path}/{state}")
    for year in years:
        quarter = os.listdir(f"{map_insstate_path}/{state}/{year}")
        for qtr in quarter:
            # print(f"{map_insstate_path}/{state}/{year}/{qtr}")
            path = f"{map_insstate_path}/{state}/{year}/{qtr}"
            with open(path, 'r') as f:
                data = json.load(f)

            # Get the actual data rows
            try:
                rows = data['data']['data']['data']  # list of [lat, lng, metric, label]
            except KeyError:
                continue  # skip if missing

            for row in rows:
                try:
                    lat = row[0]
                    lng = row[1]
                    metric = row[2]
                    label = row[3]
                except (IndexError, TypeError):
                    lat = lng = metric = label = None

                map_ins_data['State'].append(state)
                map_ins_data['Year'].append(year)
                try:
                    quarter = int(qtr.strip('.json'))
                except:
                    quarter = None
                map_ins_data['Quarter'].append(quarter)
                map_ins_data['District'].append(label)
                map_ins_data['Latitude'].append(lat)
                map_ins_data['Longitude'].append(lng)
                map_ins_data['Metric'].append(metric)

# Convert to DataFrame
map_insC_df = pd.DataFrame(map_ins_data)
map_insC_df
```

```python
'''
FOR DATABASE INSERTION OF MAP INSURANCE DATA
'''
map_insC_df.to_sql('map_insurance_country', engine, if_exists='replace', index=False)
print("✅ insuranceCountry data inserted into PostgreSQL successfully.")
```

```python
'''
For MAP INSURANCE META DATA
This code processes insurance metadata from JSON files organized by state, year, and quarter.
'''

import pandas as pd
import os
import json

map_insstate_path = r'D:\guvi project\PhonePay\pulse\data\map\insurance\country\india\state'
map_insstate_list = os.listdir(map_insstate_path)

map_ins_meta_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'DataLevel': [], 'GridLevel': [],
    'P10': [], 'P20': [], 'P30': [], 'P40': [], 'P50': [],
    'P60': [], 'P80': [], 'P90': [], 'P99_5': []
}
for state in map_insstate_list:
    years = os.listdir(f"{map_insstate_path}/{state}")
    for year in years:
        quarter = os.listdir(f"{map_insstate_path}/{state}/{year}")
        for qtr in quarter:
            # print(f"{map_insstate_path}/{state}/{year}/{qtr}")
            path = f"{map_insstate_path}/{state}/{year}/{qtr}"
            with open(path, 'r') as f:
                data = json.load(f)

            try:
                meta = data.get('data', {}).get('meta', {})
                percentiles = meta.get('percentiles', {})
            except Exception as e:
                print(f"Error accessing meta in {path}: {e}")
                meta = {}
                percentiles = {}

            map_ins_meta_data['State'].append(state)
            map_ins_meta_data['Year'].append(year)
            try:
                map_ins_meta_data['Quarter'].append(int(qtr.strip('.json')))
            except:
                map_ins_meta_data['Quarter'].append(None)

            try:
                map_ins_meta_data['DataLevel'].append(meta.get('dataLevel'))
            except:
                map_ins_meta_data['DataLevel'].append(None)

            try:
                map_ins_meta_data['GridLevel'].append(meta.get('gridLevel'))
            except:
                map_ins_meta_data['GridLevel'].append(None)

            # Each percentile with default None fallback
            map_ins_meta_data['P10'].append(percentiles.get('10.0'))
            map_ins_meta_data['P20'].append(percentiles.get('20.0'))
            map_ins_meta_data['P30'].append(percentiles.get('30.0'))
            map_ins_meta_data['P40'].append(percentiles.get('40.0'))
            map_ins_meta_data['P50'].append(percentiles.get('50.0'))
            map_ins_meta_data['P60'].append(percentiles.get('60.0'))
            map_ins_meta_data['P80'].append(percentiles.get('80.0'))
            map_ins_meta_data['P90'].append(percentiles.get('90.0'))
            map_ins_meta_data['P99_5'].append(percentiles.get('99.5'))

# Convert to DataFrame
map_insCMeta_df = pd.DataFrame(map_ins_meta_data)
map_insCMeta_df
```

```python
'''
FOR DATABASE INSERTION OF MAP INSURANCE META DATA
'''
map_insCMeta_df.to_sql('map_insurance_country_meta', engine, if_exists='replace', index=False)
print("✅ insuranceCountry data inserted into PostgreSQL successfully.")
```

```python
'''
FOR TOP INSURANCE DISTRICT AND PINCODE DATA
This code processes top insurance data from JSON files organized by state, year, and quarter.
'''

import pandas as pd
import os
import json

top_insState_path = r'D:\guvi project\PhonePay\pulse\data\top\insurance\country\india\state'
top_insState_list = os.listdir(top_insState_path)

top_ins_district_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'District': [], 'Type': [], 'Count': [], 'Amount': []
}

top_ins_pincode_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'Pincode': [],'Type': [], 'Count': [], 'Amount': []
}

for state in top_insState_list:
    years = os.listdir(f"{top_insState_path}/{state}")
    for year in years:
        quarters = os.listdir(f"{top_insState_path}/{state}/{year}")
        for qtr in quarters:
            path = f"{top_insState_path}/{state}/{year}/{qtr}"
            try:
                with open(path, 'r') as f:
                    data = json.load(f)
            except Exception as e:
                print(f"Error reading {path}: {e}")
                continue

            # ---- Extract districts ----
            try:
                districts = data.get('data', {}).get('districts', [])
                for d in districts:
                    top_ins_district_data['State'].append(state)
                    top_ins_district_data['Year'].append(year)
                    top_ins_district_data['Quarter'].append(int(qtr.strip('.json')))
                    top_ins_district_data['District'].append(d.get('entityName'))
                    metric = d.get('metric', {})
                    top_ins_district_data['Type'].append(metric.get('type'))
                    top_ins_district_data['Count'].append(metric.get('count'))
                    top_ins_district_data['Amount'].append(metric.get('amount'))
            except Exception as e:
                print(f"Error in districts for {path}: {e}")

            # ---- Extract pincodes ----
            try:
                pincodes = data.get('data', {}).get('pincodes', [])
                for p in pincodes:
                    top_ins_pincode_data['State'].append(state)
                    top_ins_pincode_data['Year'].append(year)
                    top_ins_pincode_data['Quarter'].append(int(qtr.strip('.json')))
                    top_ins_pincode_data['Pincode'].append(p.get('entityName'))
                    metric = p.get('metric', {})
                    top_ins_pincode_data['Type'].append(metric.get('type'))
                    top_ins_pincode_data['Count'].append(metric.get('count'))
                    top_ins_pincode_data['Amount'].append(metric.get('amount'))
            except Exception as e:
                print(f"Error in pincodes for {path}: {e}")

# Convert to DataFrames
top_ins_dis_df = pd.DataFrame(top_ins_district_data)
top_ins_pin_df = pd.DataFrame(top_ins_pincode_data)
print(top_ins_dis_df.head())
print(top_ins_pin_df.head())
```

```python
'''
FOR DATABASE INSERTION OF TOP INSURANCE DISTRICT DATA AND PINCODE DATA
'''
top_ins_dis_df.to_sql('top_insurance_district', engine, if_exists='replace', index=False)
print("✅ top insurance district data inserted into PostgreSQL successfully.")

top_ins_pin_df.to_sql('top_insurance_pincode', engine, if_exists='replace', index=False)
print("✅ top insurance pincode data inserted into PostgreSQL successfully.")
```

```python
'''
TOP TRANSACTION DISTRICT AND PINCODE DATA
This code processes top transaction data from JSON files organized by state, year, and quarter.
'''

import pandas as pd
import os
import json

# Base path JSON files
top_transState_path = r'D:\guvi project\PhonePay\pulse\data\top\transaction\country\india\state'
top_transState_list = os.listdir(top_transState_path)

district_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'District': [], 'Transaction_count': [], 'Transaction_amount': []
}

pincode_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'Pincode': [], 'Transaction_count': [], 'Transaction_amount': []
}

for state in top_transState_list:
    years = os.listdir(os.path.join(top_transState_path, state))
    for year in years:
        quarters = os.listdir(os.path.join(top_transState_path, state, year))
        for qtr in quarters:
            path = os.path.join(top_transState_path, state, year, qtr)
            try:
                with open(path, 'r') as f:
                    data = json.load(f)
            except Exception as e:
                print(f"Failed to load {path}: {e}")
                continue

            # Extract districts
            try:
                districts = data.get("data", {}).get("districts", [])
                for d in districts:
                    metric = d.get('metric', {})
                    district_data['State'].append(state)
                    district_data['Year'].append(year)
                    district_data['Quarter'].append(int(qtr.strip('.json')))
                    district_data['District'].append(d.get('entityName'))
                    district_data['Transaction_count'].append(metric.get('count'))
                    district_data['Transaction_amount'].append(metric.get('amount'))
            except Exception as e:
                print(f"Error reading districts from {path}: {e}")

            # Extract pincodes
            try:
                pincodes = data.get("data", {}).get("pincodes", [])
                for p in pincodes:
                    metric = p.get('metric', {})
                    pincode_data['State'].append(state)
                    pincode_data['Year'].append(year)
                    pincode_data['Quarter'].append(int(qtr.strip('.json')))
                    pincode_data['Pincode'].append(p.get('entityName'))
                    pincode_data['Transaction_count'].append(metric.get('count'))
                    pincode_data['Transaction_amount'].append(metric.get('amount'))
            except Exception as e:
                print(f"Error reading pincodes from {path}: {e}")

# Convert to DataFrames
top_trans_dis_df = pd.DataFrame(district_data)
top_trans_pin_df = pd.DataFrame(pincode_data)

# Preview
print("District Data:")
print(top_trans_dis_df.head())

print("\nPincode Data:")
print(top_trans_pin_df.head())
```

```python
'''
FOR DATABASE INSERTION OF TOP TRANSACTION DISTRICT DATA AND PINCODE DATA
'''
top_trans_dis_df.to_sql('top_transaction_district', engine, if_exists='replace', index=False)
print("✅ top transaction district data inserted into PostgreSQL successfully.")

top_trans_pin_df.to_sql('top_transaction_pincode', engine, if_exists='replace', index=False)
print("✅ top transaction pincode data inserted into PostgreSQL successfully.")
```

```python
'''
USER DISTRICT AND PINCODE DATA
This code processes user data from JSON files organized by state, year, and quarter.
'''

import pandas as pd
import os
import json

# Path to top/user/state JSON data
top_user_state_path = r'D:\guvi project\PhonePay\pulse\data\top\user\country\india\state'
top_user_state_list = os.listdir(top_user_state_path)

# Initialize data holders
user_district_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'District': [], 'RegisteredUsers': []
}

user_pincode_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'Pincode': [], 'RegisteredUsers': []
}

for state in top_user_state_list:
    years = os.listdir(os.path.join(top_user_state_path, state))
    for year in years:
        quarters = os.listdir(os.path.join(top_user_state_path, state, year))
        for qtr in quarters:
            path = os.path.join(top_user_state_path, state, year, qtr)
            try:
                with open(path, 'r') as f:
                    data = json.load(f)
            except Exception as e:
                print(f"Failed to load {path}: {e}")
                continue

            # Extract Districts
            try:
                districts = data.get("data", {}).get("districts", [])
                for d in districts:
                    user_district_data['State'].append(state)
                    user_district_data['Year'].append(year)
                    user_district_data['Quarter'].append(int(qtr.strip('.json')))
                    user_district_data['District'].append(d.get('name'))
                    user_district_data['RegisteredUsers'].append(d.get('registeredUsers'))
            except Exception as e:
                print(f"Error reading districts from {path}: {e}")

            # Extract Pincodes
            try:
                pincodes = data.get("data", {}).get("pincodes", [])
                for p in pincodes:
                    user_pincode_data['State'].append(state)
                    user_pincode_data['Year'].append(year)
                    user_pincode_data['Quarter'].append(int(qtr.strip('.json')))
                    user_pincode_data['Pincode'].append(p.get('name'))
                    user_pincode_data['RegisteredUsers'].append(p.get('registeredUsers'))
            except Exception as e:
                print(f"Error reading pincodes from {path}: {e}")

# Convert to DataFrames
top_user_dis_df = pd.DataFrame(user_district_data)
top_user_pin_df = pd.DataFrame(user_pincode_data)

# Preview
print("User District Data:")
print(top_user_dis_df.head())

print("\nUser Pincode Data:")
print(top_user_pin_df.head())

```

```python
'''
FOR DATABASE INSERTION OF TOP USER DISTRICT DATA AND PINCODE DATA
'''
top_user_dis_df.to_sql('top_user_district', engine, if_exists='replace', index=False)
print("✅ top user district data inserted into PostgreSQL successfully.")

top_user_pin_df.to_sql('top_user_pincode', engine, if_exists='replace', index=False)
print("✅ top user pincode data inserted into PostgreSQL successfully.")
```

```python
'''
TO AGGREGATED INSURANCE DATA
This code processes aggregated insurance data from JSON files organized by state, year, and quarter.
'''

import pandas as pd
import os
import json

# Base path to aggregated insurance data
agg_ins_path = r"D:\guvi project\PhonePay\pulse\data\aggregated\insurance\country\india\state"
agg_ins_states = os.listdir(agg_ins_path)

agg_ins_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'From': [], 'To': [],
    'Type': [], 'Transaction_count': [], 'Transaction_amount': []
}

for state in agg_ins_states:
    years = os.listdir(os.path.join(agg_ins_path, state))
    for year in years:
        quarters = os.listdir(os.path.join(agg_ins_path, state, year))
        for qtr in quarters:
            path = os.path.join(agg_ins_path, state, year, qtr)
            try:
                with open(path, 'r') as f:
                    data = json.load(f)
            except Exception as e:
                print(f"Failed to load {path}: {e}")
                continue

            from_ts = data.get("data", {}).get("from", None)
            to_ts = data.get("data", {}).get("to", None)
            transaction_data = data.get("data", {}).get("transactionData", [])

            for txn in transaction_data:
                name = txn.get("name", None)
                instruments = txn.get("paymentInstruments", [])

                for instr in instruments:
                    count = instr.get("count", None)
                    amount = instr.get("amount", None)

                    agg_ins_data['State'].append(state)
                    agg_ins_data['Year'].append(year)
                    agg_ins_data['Quarter'].append(int(qtr.strip('.json')))
                    agg_ins_data['From'].append(from_ts)
                    agg_ins_data['To'].append(to_ts)
                    agg_ins_data['Type'].append(name)
                    agg_ins_data['Transaction_count'].append(count)
                    agg_ins_data['Transaction_amount'].append(amount)

# Convert to DataFrame
agg_ins_df = pd.DataFrame(agg_ins_data)
agg_ins_df
# Show preview
# print(agg_ins_df.head())
```

```python
import os
import json
import pandas as pd

agg_user_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'RegisteredUsers': [], 'AppOpens': []
}

agg_user_device_data = {
    'State': [], 'Year': [], 'Quarter': [],
    'Brand': [], 'Count': [], 'Percentage': []
}
agg_user_path = r'D:\guvi project\PhonePay\pulse\data\aggregated\user\country\india\state'
state_list = os.listdir(agg_user_path)
for state in state_list:
    years = os.listdir(os.path.join(agg_user_path, state))
    for year in years:
        quarters = os.listdir(os.path.join(agg_user_path, state, year))
        for qtr in quarters:
            path = os.path.join(agg_user_path, state, year, qtr)
            try:
                with open(path, 'r') as f:
                    data = json.load(f)
            except Exception as e:
                print(f"Error loading {path}: {e}")
                continue
            
            agg = data.get('data', {}).get('aggregated', {})

            # Extract aggregated user data
            agg_user_data['State'].append(state)
            agg_user_data['Year'].append(year)
            agg_user_data['Quarter'].append(int(qtr.strip('.json')))
            agg_user_data['RegisteredUsers'].append(agg.get('registeredUsers'))
            agg_user_data['AppOpens'].append(agg.get('appOpens'))

            # Extract users by device
            users = data.get('data', {}).get('usersByDevice')
            if isinstance(users, list):
                for u in users:
                    # print(u)
                    agg_user_device_data['State'].append(state)
                    agg_user_device_data['Year'].append(year)
                    agg_user_device_data['Quarter'].append(int(qtr.strip('.json')))
                    agg_user_device_data['Brand'].append(u.get('brand'))
                    agg_user_device_data['Count'].append(u.get('count'))
                    agg_user_device_data['Percentage'].append(u.get('percentage'))
            else:
                agg_user_device_data['State'].append(state)
                agg_user_device_data['Year'].append(year)
                agg_user_device_data['Quarter'].append(int(qtr.strip('.json')))
                agg_user_device_data['Brand'].append(None)
                agg_user_device_data['Count'].append(None)
                agg_user_device_data['Percentage'].append(None)

# Convert to DataFrames
agg_user_df = pd.DataFrame(agg_user_data)
agg_device_user_df = pd.DataFrame(agg_user_device_data)


# Preview
print("Aggregated User Data:")
print(agg_user_df.head())

print("\nDevice User Data:")
print(agg_device_user_df.head())
```
